# Dice Rolling Simulator
##Problem Statement

In many games, a physical dice is used to generate a random number between 1 and 6. This project aims to simulate the behavior of a real six-sided dice using Python.

The program should allow the user to roll the dice and display a random number from 1 to 6. The user can continue rolling the dice multiple times until they choose to quit the game.

##Objectives
Learn how to use Python's random module.
Understand user input and output.
Practice loops and conditional statements.
Simulate a real-world object using programming.

##Requirements
Generate a random number between 1 and 6 whenever the dice is rolled.
Display the result to the user.
Ask the user whether they want to roll again.
Continue the game until the user chooses to exit.
Display a farewell message before terminating the program.

```python
import random

dice_faces = {
    1: '''
+-------+
|       |
|   ●   |
|       |
+-------+
''',

    2: '''
+-------+
| ●     |
|       |
|     ● |
+-------+
''',

    3: '''
+-------+
| ●     |
|   ●   |
|     ● |
+-------+
''',

    4: '''
+-------+
| ●   ● |
|       |
| ●   ● |
+-------+
''',

    5: '''
+-------+
| ●   ● |
|   ●   |
| ●   ● |
+-------+
''',

    6: '''
+-------+
| ●   ● |
| ●   ● |
| ●   ● |
+-------+
'''
}

print("🎲 Welcome to the Dice Rolling Simulator!")

play = input("Do you want to roll a die? (Y/N): ").strip().lower()

if play == 'y':
    no_of_rolls = int(input("How many times do you want to roll the die? "))

    for roll in range(1, no_of_rolls + 1):
        die = random.randint(1, 6)

        print(f"\nRoll #{roll}")
        print(f"You rolled a {die}")
        print(dice_faces[die])

    print("Finished rolling!")

elif play == 'n':
    print("Thanks for playing!")
    print("Bye!")

else:
    print("Invalid choice!")
```
